```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv("Data/original/verbraucherpreisindex.csv")

# Convert the 'DATE' column to datetime objects
df["DATE"] = pd.to_datetime(df["DATE"], format="%m-%d-%y")

# Save the DataFrame with the converted DATE column
# The DATE column in the output CSV will be in a standard format (e.g., YYYY-MM-DD or YYYY-MM-DD HH:MM:SS)
df.to_csv("Data/monthly/verbraucherpreisindex monthly.csv", index=False)
df
```
