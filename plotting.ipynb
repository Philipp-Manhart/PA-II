```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the dataset
df = pd.read_csv("/home/philipp/Code/BigDataBI/Data/monthly/all data.csv")

# Convert 'DATE' column to datetime objects
df["DATE"] = pd.to_datetime(df["DATE"])

# Get a list of all columns except 'DATE'
columns_to_plot = df.columns.drop("DATE")

# Create a plot for each column
for column in columns_to_plot:
    plt.figure(figsize=(12, 6))
    plt.plot(df["DATE"], df[column], marker="o", linestyle="-")
    plt.title(f"{column} over Time")
    plt.xlabel("Date")
    plt.ylabel(column)
    plt.grid(True)

    # Format the x-axis to show dates nicely
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    plt.gca().xaxis.set_major_locator(
        mdates.MonthLocator(interval=6)
    )  # Show a tick every 6 months
    plt.gcf().autofmt_xdate()  # Rotate date labels for better readability

    plt.show()
```
